```python
import matplotlib.pyplot as plt
```

```python
values=[1,2,3,4,5,6,7]
square=[1,4,9,16,25,36,49]
cube=[1,8,27,64,125,216,343]
plt.plot(values,square,linewidth=3,c="red")
plt.plot(values,cube,linewidth=4,dash_capstyle="round",c="blue")
plt.title("square numbers",fontsize=30)
plt.xlabel("Values",fontsize=20)
plt.ylabel("Values",fontsize=20)
plt.tick_params(axis="both",labelsize="15")
####################################
```

```python
plt.scatter(values,square,s=20)
plt.show()
```

```python
x_values=list(range(1,8))
x_values
```

```python
y_values=list(i**2 for i in x_values)
y2_values=[i**3 for i in x_values]
print(y_values)
print(y2_values)
```

```python
plt.scatter(x_values,y_values,s=20,c="red")
plt.scatter(x_values,y2_values,s=30,c="blue")
```

```python
plt.scatter(x_values,y2_values,s=30,c=y2_values,cmap=plt.cm.Blues)
```
