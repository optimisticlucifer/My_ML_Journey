```python
#plotting points with scatter()
import matplotlib.pyplot as plt

#matplotlib.pyplot.scatter(1,86)
#matplotlib.pyplot.show()
```

```python
#plotting series of Points

x_values = [1,2,3,4,5]
y_values = [1,4,9,16,25]

plt.scatter(x_values , y_values , s = 100)
plt.show()
```

Calculating Data Automatically

```python

print(x_values)
y_values =[x**2 for x in x_values]
print(y_values)
#x**3 for x in x_values
```

```python
x_values = list(range(1 , 10))
x_values
```

```python
x_values = list(range(1 , 15))
print(x_values)
y_values = [x**3 for x in x_values]

y_values2 = [x**2 for x in x_values]
print(y_values)
plt.scatter(x_values , y_values , s =40)
plt.scatter(x_values , y_values2 , s = 70)
plt.show()
```

```python
#Changing the color
plt.scatter(x_values , y_values , c = 'red' , s = 40)
plt.show()
```

```python
plt.scatter(x_values , y_values , c = ( 0.1, 0.9, 0.1) , s = 40)
plt.show()

#values closer to 0 will produce dark colors
```

```python
#Using ColorMap

plt.scatter(x_values , y_values , c = y_values , cmap = plt.cm.Blues , s = 40)
plt.show()

```
